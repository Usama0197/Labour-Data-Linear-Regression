###  Separate Numerical and Categorical Variables
  Numerical Variables:

```python
df_num = df[['Age','Nodeg','Earnings_1974','Earnings_1975','Earnings_1978']]
```

Categorical Variables: The Variables which are having category should be converted to dummy variables.

The Linear Regression will only accept numbers so we will make dummy variables.

### Creation of dummy variables

```python
df_dummies = pd.get_dummies(df[['Race','Hisp','MaritalStatus','Eduacation']])
```

```python
df_dummies.head()
```

### Combine Numerical and Dummy variables

```python
df_combined = pd.concat([df_num,df_dummies],axis=1)
```

```python
df_combined.head()
```

# Step 3 : Splitting the data in Training and Test set

 Using sklearn we split 70% of our data into training and rest in Test set.
 
 Setting random_state will give the same Training and Test set everytime on running the code.

```python
# putting feature variable to x
X = df_combined[['Age','Earnings_1974','Earnings_1975','Race_NotBlack','Race_black',
                        'Hisp_NotHispanic','Hisp_hispanic','MaritalStatus_Married',
                        'MaritalStatus_NotMarried','Eduacation_HighSchool','Eduacation_Intermediate',
                        'Eduacation_LessThanHighSchool','Eduacation_PostGraduate','Eduacation_graduate']]

# putting response variable to y
y = df['Earnings_1978']
```

```python
# random_state is the seed used by the random number generator. It can be any integer.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=100)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

# Step 4 : Performing Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
# Representing LinearRegression as lr(Creating LinearRegression Object)
lr = LinearRegression()
```

```python
# fit the model to the training data
lr.fit(X_train, y_train )
```

# Step 5 : Model Evaluation

```python
# print the intercept
print(lr.intercept_)
```

```python
# Let's see the coefficient
coeff_df = pd.DataFrame(lr.coef_,X_test.columns,columns=['Coefficient'])
coeff_df
```

# Step 6 : Predictions

```python
# Making predictions using the model
y_pred = lr.predict(X_test)
```

# Step 7 : Model Performance Metrics

### Coefficient of Determination (R square)

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test,y_pred)
r_squared = r2_score(y_test,y_pred)
```

```python
from math import sqrt
rmse = sqrt(mse)
```

```python
print('Mean Squared Error :',mse)
print('Root_Mean_Squared_Error :',rmse)
print('r_square_value :',r_squared)
```
